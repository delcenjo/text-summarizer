```python
from textsummarizer import summarize, key_points

text = '''Hugging Face Spaces let you host machine learning demos directly from a
repository. A Space is a git repo with an app file and a requirements file; when you
push, the platform builds an environment and runs the app, giving you a public URL.
Spaces support Gradio and Streamlit out of the box, and the free tier runs on CPU,
which is enough for small models. This makes it easy to share a working demo of a
model with anyone, without managing servers yourself.'''

print(summarize(text))
```

## Key points

```python
for p in key_points(summarize(text)):
    print('-', p)
```
